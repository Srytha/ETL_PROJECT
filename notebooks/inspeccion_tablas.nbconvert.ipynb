```python
import psycopg2
import pandas as pd
```

```python
try:
    conn = psycopg2.connect(**config)
    
    query = """
        SELECT table_name 
        FROM information_schema.tables 
        WHERE table_schema = 'public'
        ORDER BY table_name;
    """
    df_tablas = pd.read_sql(query, conn)

    print(f"\nTotal de tablas encontradas: {len(df_tablas)}")
    print("\nLista de tablas:")
    for i, tabla in enumerate(df_tablas['table_name'], 1):
        print(f"  {i}. {tabla}")

except Exception as e:
    print(f"Error de conexion: {e}")
```
