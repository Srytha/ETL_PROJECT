```python
import numpy as np
import pandas as pd
import yaml
from sqlalchemy import create_engine
```

#### Conexion con las bases de datos

```python
with open('../config.yml', 'r') as f:
    config = yaml.safe_load(f)
    config_co = config['SOURCE_DB']
    config_etl = config['ETL_PRO']
    
url_co = (f"{config_co['drivername']}://{config_co['user']}:{config_co['password']}@{config_co['host']}:"
          f"{config_co['port']}/{config_co['dbname']}")
url_etl = (f"{config_etl['drivername']}://{config_etl['user']}:{config_etl['password']}@{config_etl['host']}:"
           f"{config_etl['port']}/{config_etl['dbname']}")

engine_origen = create_engine(url_co)
engine_destino = create_engine(url_etl)
```

```python
estado = pd.read_sql_table("mensajeria_estado", engine_origen)
```

```python
estado.head()
```

```python
estado.rename(columns={
    'id': 'id_estado'
}, inplace=True)
```

```python
columnas_necesarias = ["id_estado", "nombre", "descripcion"]
dim_estado= estado[columnas_necesarias].copy()
dim_estado.head()
```

```python
dim_estado.to_sql('dim_estado',con=engine_destino,if_exists='replace',index_label='key_dim_estado')
```
