```python
import numpy as np
import pandas as pd
import yaml
from sqlalchemy import create_engine
```

```python
config = {
    'host': 'localhost',       
    'port': 
    'database': 'mensajeria',  
    'user': 'postgres',            
    'password': 
}
```

```python
with open("../config.yml", "r") as f:
    config = yaml.safe_load(f)

    config_co = config['SOURCE_DB']
    config_etl = config['ETL_PRO']

url_co = f"{config_co['drivername']}://{config_co['user']}:{config_co['password']}@{config_co['host']}:{config_co['port']}/{config_co['dbname']}"
url_etl = f"{config_etl['drivername']}://{config_etl['user']}:{config_etl['password']}@{config_etl['host']}:{config_etl['port']}/{config_etl['dbname']}"


engine_origen = create_engine(url_co)
engine_destino = create_engine(url_etl)
```

```python
dim_mensajero = pd.read_sql_table("clientes_mensajeroaquitoy", engine_origen)
dim_mensajero.head()
```

```python
dim_mensajero.rename(
    columns={
        "id": "id_mensajero",
    },
    inplace=True,
)
```

```python
dim_mensajero.columns.tolist()
```

```python
columnas_necesarias = ["id_mensajero","activo"]
dim_mensajero = dim_mensajero[columnas_necesarias].copy()
dim_mensajero.head()
```

```python
dim_mensajero.to_sql(
    "dim_mensajero",
    con=engine_destino,
    if_exists="replace",
    index_label="key_dim_mensajero",
)
```
